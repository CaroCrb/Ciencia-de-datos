# Prediciendo estrellas 🌟

El dataset que se usará será uno que nos predice los tipos de estrellas y los ubica en el diagrama HR o Diagrama de Hertzsprung-Russell.

Me gustaría analizar este conjunto de datos para saber qué tipo de estrellas son las más comunes y si en realidad, a cada tipo de estrella le pertenece algún color en específico o no. De igual forma, pretendo comparar los resultados con el diagrama H-R para comprobar lo obtenido en el análisis a lo largo del trabajo.

Según la literatura, el tipo más común son las estrellas frías, por lo que veremos si eso es lo que sucede con el dataset elegido.


```python
#Importamos primero el dataset
data = pd.read_csv('C:/Users/karol/Desktop/Computación/6_class_csv.csv')
```

```python
data
```

```python
datos=data.Star_type.value_counts() #Tabla1
datos
```

```python
data.Star_color.value_counts() #Tabla2
```

Podemos notar en la Tabla1 que hay 5 tipos de estrellas y que en este dataset hay 40 estrellas de cada uno; sin embargo, vemos en la Tabla2 que no necesariamente todas las que pertenecen a una misma categoría tienen el mismo color (pues hay 19 distintos)

```python
sns.scatterplot(data=data, x="Star_type", y="Star_color", alpha=0.2); #Gráfica1
```

Vemos entonces en la gráfica 1 que las estrellas del tipo 2, 3, 4 y 5 pueden ser de color azul, y a su vez, las estrellas del tipo 0, 1, 4 y 5 también pueden ser de color rojo (y como vimos en la segunda tabla, la mayoría de estrellas son de ese color), color que representa a las estrellas frías.


```python
espectros=data.Spectral_Class.value_counts() #Tabla3
espectros
```

En esta tercer tabla, vemos que las estrellas tipo M son las más abundantes, siguiendo las estrellas tipo B y ambas, con mucha diferencia respecto a las menos comunes que son las estrellas de clase espectral G. 

En la gráfica 2 tenemos una vista general del valor de la temperatura (en aumento) con la luminosidad de las estrellas y el tipo de éstas, coincidiendo con la gráfica 1, ya que esa nos dice que las estrellas tipo 3 pueden llegar a ser muy azules (calientes), y en esta gráfica vemos ese tipo de estrella hasta en temperaturas de 40000K.


```python
sns.scatterplot(data=data, x="Temperature_K", y="Luminosity", alpha=0.3, hue="Star_type", 
    palette="Spectral"); #Gráfica 2
```

Respecto a las magnitudes absolutas, veamos la siguiente gráfica (Gráfica 3):

```python
sns.violinplot(data=data, y="Absolute_magnitude", x="Spectral_Class") #Gráfica 3
```

Podemos notar en la gráfica 3 que las estrellas tipo M, en general, puede tomar distintos tamaños (no tiene alguno en específico) y lo mismo con todos los distintos tipos...menos con las G's, ellas son muy pequeñitas⭐

Por otra parte, veremos ahora el diagrama H-R

# Conclusiones

El análisis de los datos nos arrojó que las estrellas tipo M son las más comunes y se caracterizan por ser estrellas frías (respecto al resto) y con un color rojizo. 

Sabemos además, que el analisis fue acertado pues coincidió con el diagrama H-R que es aquel que nos sitúa a las estrellas dependiendo la temperatura, luminosidad, tamaño, tipo espectral, etc.
